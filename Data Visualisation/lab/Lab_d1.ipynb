```python
import seaborn as sns
import matplotlib.pyplot as plt

import os
os.getcwd()
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

sns.set_style("whitegrid")

plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]

plt.plot(x1, label = 'Group A')
plt.plot(x2, label = 'Group B')

plt.legend()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure()
x1 = [10,20,5,40,8]
x2 = [30,43,9,7,20]

with sns.axes_style('ticks'):
    plt.plot(x1, label = 'group A')
    plt.plot(x2, label = 'Group B')

sns.despine(fig = None, ax = None, top = False, right = False, left = False, bottom = False, offset = None,trim = False)    

plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

movie = pd.read_csv("../DAVI Dataset/movie_score.csv", index_col = 0)

sns.set(style = "darkgrid")
sns.set_context("paper")

movie = pd.melt(movie, id_vars = ["MovieTitle"], var_name=['Type']. value_vars = ["Tomatometer","AudienceScore"], value_name = 'Count')

barplot = sns.barplot(x="MovieTitle", y="Count", hue= "Type", data = movie, hue_order = ["AudienceScore","TomatoScore"])
```
