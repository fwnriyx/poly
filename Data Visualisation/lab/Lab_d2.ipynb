```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

score = pd.read_csv("../DAVI Dataset/scores.csv")
score = pd.melt(score, value_vars=['group_a', 'group_b', 'group_c', 'group_d'], 
             var_name='Groups', value_name='IQ score')

print(score)
%matplotlib inline
sns.violinplot(x="Groups", y="IQ score", data= score)
plt.title("IQ scores for different test groups")
plt.show()
```

```python
df = pd.read_csv("../DAVI Dataset/youtube.csv")
df = pd.melt(df, id_vars =["channels"], value_vars=['subs','views'], 
             var_name='Type', value_name ="Subs in millions")


g = sns.FacetGrid(df, col = 'Type', sharex = False,
                    height=17, hue_kws={'color': ['b','g']}, hue = 'Type')
g.map_dataframe(sns.barplot, 'Subs in millions','channels')


plt.show()
```

```python
import squarify

colors = sns.light_palette("green",4)

usage = pd.read_csv("../DAVI Dataset/water_usage.csv")
print(usage.head())
prec = '(' + usage.Percentage.astype(str) + '%)'
squarify.plot(sizes = usage.Percentage,
              label= usage.Usage, value = prec, alpha = 0.6, color = colors)

plt.show()
```
