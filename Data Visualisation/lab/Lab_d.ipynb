```python
#A
#1. Read iris data using pandas
#2. Create a scatter plot to analyze the relationship between sepal length and
# sepal width
#3. Background is white
#4. Use jointplot() function

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = pd.read_csv("../DAVI Dataset/iris.csv")
print(iris.head())
sns.set(style = "white")

sns.jointplot(data = iris, x="SepalLengthCm", y ="SepalWidthCm")
plt.show()
```

```python
sns.jointplot(data = iris, x="SepalLengthCm", y= "SepalWidthCm", kind = "kde", fill = True)
sns.set(style="white")
plt.show()
```

```python
data = pd.read_csv("../DAVI Dataset/iris.csv", index_col = 0)

sns.pairplot(data, hue = "Species")

plt.show()
```

```python
tips = pd.read_csv("../DAVI Dataset/tips.csv")
print(tips.head())
g = sns.FacetGrid(tips, col="time")
g.map(sns.violinplot, "sex", "total_bill", "smoker", split = True)


plt.show()
```

```python
sns.lmplot(
    data=tips, x="total_bill", y="tip", x_bins = 5)
plt.show()
```

```python
day_sex_table = pd.crosstab(tips['day'], tips['sex'])

sns.heatmap(day_sex_table, cmap="BuGn", cbar_kws={'label': 'My Color Bar', 'orientation': 'horizontal'}, linewidths=0.3, annot = True)

plt.show()
```

```python
import numpy as np
sns.heatmap(tips.corr(), annot = True, cmap="BuGn", 
cbar_kws={'label': 'My Color Bar', 'orientation': 'horizontal'})
sns.heatmap(tips.corr(), annot = True, cmap="BuGn", cbar = False)

plt.show()
```

```python
sns.heatmap(tips.corr(), annot = True, cmap="BuGn", 
cbar_kws={'label': 'My Color Bar', 'orientation': 'horizontal'}cbar = True, 
mask = np.triu(tips.corr()))
plt.show()
```

```python
sns.heatmap(tips.corr().sort_values(by='total_bill',ascending=False), 
annot = True, cmap="BuGn", cbar = True)
plt.show()
```
